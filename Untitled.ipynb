```python
import numpy as np
import pandas as pd

# Sample banking data
np.random.seed(42)
n_customers = 10
loan_amount = np.random.randint(1000, 10000, n_customers)
d0 = np.random.uniform(0.01, 0.10, n_customers)  # base default probability
k = 0.5  # sensitivity of default to interest rate

# Optimal interest rate
optimal_rate = (1 - d0) / (2 * k)

# Calculate optimized NII
NII_optimal = loan_amount * optimal_rate * (1 - (d0 + k*optimal_rate))

data = pd.DataFrame({
    'LoanAmount': loan_amount,
    'BaseDefault': d0,
    'OptimalRate': optimal_rate,
    'NII_Optimal': NII_optimal
})

data
```

```python
import matplotlib.pyplot as plt

rate_range = np.linspace(0.01, 1.0, 100)
nii_total = [np.sum(loan_amount * r * (1 - (d0 + k*r))) for r in rate_range]

plt.plot(rate_range, nii_total)
plt.axvline(np.mean(optimal_rate), color='red', linestyle='--', label='Average Optimal Rate')
plt.xlabel('Interest Rate')
plt.ylabel('Total NII')
plt.title('Sensitivity of NII to Interest Rate')
plt.legend()
plt.grid(True)
plt.show()
```
